```python
import networkx as nx
from matplotlib import pyplot as plt
import numpy as np
import random
```

## (Problem 3a):

```python
nodes = range(10)
G = nx.Graph()
G.add_nodes_from(nodes)

#randomly generate 15 edges between the nodes
num_edges = 0
while num_edges < 15:
    n1, n2 = np.random.randint(0, 10), np.random.randint(0, 10)

    if (n1, n2) not in G.edges and (n2, n1) not in G.edges:
        num_edges += 1
        G.add_edge(n1, n2)

pos = nx.spring_layout(G, seed=0) 
nx.draw(G, pos=pos, with_labels=True)
```

## (Problem 3b):

```python
#selected the edges and nodes to highlight
random_nodes = np.random.randint(0, 10, size=2)
random_edges = random.sample(list(G.edges), k=2)

node_colors = {node: "tab:blue" if node not in random_nodes else "lime" for node in G.nodes}
edge_colors = {edge: "black" if edge not in random_edges else "red" for edge in G.edges}

pos = nx.spring_layout(G, seed=0) 
nx.draw(G, pos=pos, node_color=node_colors.values(), edge_color=edge_colors.values(), with_labels=True)
plt.show()
```

## (Problem 3c):

```python
#problem 3c
edges = G.edges
nodes = G.nodes
G = nx.DiGraph()
G.add_nodes_from(nodes)
G.add_edges_from(edges)

nx.draw(G, pos=pos, node_color=node_colors.values(), edge_color=edge_colors.values(), with_labels=True)
plt.show()
```
